```python
import pandas as pd
df=pd.read_csv('bank_loan_risk_dataset.csv')
df
```

Q1 Average loan amount by employment type

```python
a=df.groupby('employment_type')['loan_amount'].mean()
a
```

Q2.Default rate by employment type

```python
df['is_default'] = df['loan_status'] == 'Default'
a2=df.groupby('employment_type')['is_default'].mean()*100
a2
```

Q3.Customer have below 500 credit score

```python
a3=df[df['credit_score']<500]
a3
```

Q4.Age group have highest loan amount

```python
df.groupby('age')['loan_amount'].sum().sort_values(ascending=False).head(1)
```

Q5.Average income of approved loan customer

```python
# a5=df[df['loan_status']=='Approved']
# a5['annual_income'].mean()
df[df['loan_status']=='Approved']['annual_income'].mean()
```
